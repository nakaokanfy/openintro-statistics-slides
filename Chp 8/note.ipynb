### 練習問題 8.15

男性が次の条件の女性と結婚しているとき, 夫と妻の年齢の相関はどうなるだろうか.

- (a) 3 歳年下の場合.
- (b) 2 歳年上の場合.
- (c) 年齢が半分の場合.

#### 解答

それぞれのパートにおいて、夫の年齢を妻の年齢の一次関数として書くことができる。

- (a) 年齢H = 年齢W + 3
- (b) 年齢H = 年齢W - 2
- (c) 年齢H = 2 × 年齢W 

傾きは正であり、これらは完全な一次関係なので、相関は3つの部分すべてでちょうど1になる。この解を理解する別の方法は、模擬データセット、例えば26歳、27歳、28歳、29歳、30歳の女性5人を作り、各パートの各妻の夫の年齢を求め、散布図を作成することである。

```python

import pandas as pd
import matplotlib.pyplot as plt

# 妻の年齢データセット
wife_ages = [26, 27, 28, 29, 30]

# (a) 年齢H = 年齢W + 3
husband_ages_a = [age + 3 for age in wife_ages]

# (b) 年齢H = 年齢W - 2
husband_ages_b = [age - 2 for age in wife_ages]

# (c) 年齢H = 2 × 年齢W
husband_ages_c = [2 * age for age in wife_ages]

# データフレーム作成
data = {
  'Wife Age': wife_ages,
  'Husband Age (a)': husband_ages_a,
  'Husband Age (b)': husband_ages_b,
  'Husband Age (c)': husband_ages_c
}
df = pd.DataFrame(data)

# 散布図の作成
plt.figure(figsize=(15, 5))

# (a) の散布図
plt.subplot(1, 3, 1)
plt.scatter(df['Wife Age'], df['Husband Age (a)'], color='blue')
plt.plot(df['Wife Age'], df['Husband Age (a)'], color='blue', linestyle='--')
plt.title('(a) Husband Age = Wife Age + 3')
plt.xlabel('Wife Age')
plt.ylabel('Husband Age')

# (b) の散布図
plt.subplot(1, 3, 2)
plt.scatter(df['Wife Age'], df['Husband Age (b)'], color='green')
plt.plot(df['Wife Age'], df['Husband Age (b)'], color='green', linestyle='--')
plt.title('(b) Husband Age = Wife Age - 2')
plt.xlabel('Wife Age')
plt.ylabel('Husband Age')

# (c) の散布図
plt.subplot(1, 3, 3)
plt.scatter(df['Wife Age'], df['Husband Age (c)'], color='red')
plt.plot(df['Wife Age'], df['Husband Age (c)'], color='red', linestyle='--')
plt.title('(c) Husband Age = 2 × Wife Age')
plt.xlabel('Wife Age')
plt.ylabel('Husband Age')

# グラフの表示
plt.tight_layout()
plt.show()
```
